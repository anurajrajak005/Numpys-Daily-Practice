Q.3 Write a code to convert 1D array into 3D.And find it’s transpose
matrix.
Input: A=[1,2,3,4,5,5,6,7,8,9,10,12]


```python
A = np.array([1,2,3,4,5,5,6,7,8,9,10,12])
arr3d = A.reshape(2, 2, 3)
print("3D Array:")
print(arr3d)

print("\nTranspose:")
print(arr3d.transpose())
```

Q.4 Create 4 ×4 Identity Matrix,and replace the diagonal value with 12.

Q.9 Write a code to find odd number in given 1D array.
Input: A=[20,3,7,80,89]

```python
a=np.array([20,3,7,80,89])
b = a[a%2 != 0]
print(b)
```

Q.10 Write a code to replace 0 by 99 in identity matrix.
