## # Concatenate 

```python
import numpy as np

a = np.array([[1, 2]])
b = np.array([[3, 4]])

# Concatenate along rows (Axis 0)
row_joined = np.concatenate((a, b), axis=0)
# Result: 
# [[1, 2],
#  [3, 4]]

# Concatenate along columns (Axis 1)
col_joined = np.concatenate((a, b), axis=1)
# Result: [[1, 2, 3, 4]]
print(col_joined)
```
