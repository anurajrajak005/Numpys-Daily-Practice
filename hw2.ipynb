### 10. Remove common items from array
 Input A1=[1,2,3,4,5]
 A2=[1,4,7,8,5]
 Output=[2,3,7,8]

```python
A1=[1,2,3,4,5] 
A2=[1,4,7,8,5]
A3 = np.setxor1d(A1, A2)
print(A3)
```

### 11. Get the positions where elements of array a and b match
Input A1=[1,2,3,4,5]
A2=[1,7,8,4,5]
Output=array([0,3,4])


```python
A1=np.array([1,2,3,4,5])
A2=np.array([1,7,8,4,5])
C = np.where(A1 == A2)
print(C[0])
```

### 13 Delete second column from given array.
Input A1=[[1,2,3,4,5],[0,8,7,9,6]]

```python
A1=[[1,2,3,4,5],[0,8,7,9,6]]
A1 = np.delete(A1,2,axis=1)
print(A1)
```
